You'd want this function to:

```python
sub = prop_gb[prop_gb['London Borough'] == 'Barnet']
question = sub['Year'] == 2018
question2 = (sub['Average Price'][sub['Year'] == 2018])
print(question)
print('BREAK')
print(question2)
```

```python
# slice out 1998 data and 2018 data and then see which has highest % change 

def create_price_ratio(df_sub, start_year, end_year):
    #define start year row
    ystart = float(df_sub['Average Price'][df_sub['Year'] == start_year])
    
    #define end year row
    yend = float(df_sub['Average Price'][df_sub['Year'] == end_year]) 
    
    #divide end year avg by start year avg
    ratio = [yend / ystart] #this has to be in sqr brackets or i can't turn it in to a df later? why?

    return ratio

create_price_ratio(prop_gb[prop_gb['London Borough'] == 'Camden'], 1998, 2018)
```

```python
prop_gb['London Borough'].unique()
```

```python
#empty dict to feed new values to
df_w_ratio = {}
```

```python
#for each borough (that's why you need unique)
for b in prop_gb['London Borough'].unique():#why do I have to use .unique?
    #create parameter to feed to our UDF, df_sub = borough 
    borough = prop_gb[prop_gb['London Borough'] == b]
    #add new column (new key:value pr, borough:ratio) in df_w_ratio 
    df_w_ratio[b] = create_price_ratio(borough, 1998, 2018)

print(df_w_ratio)
```
