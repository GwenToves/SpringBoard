```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
       SELECT x.name, 
ROUND(SUM(CASE
WHEN memid = 0 THEN (x.guestcost * x.slots)
WHEN memid <> 0 THEN (x.membercost * x.slots) END), 2) AS total_revenue  
FROM (SELECT * FROM Bookings INNER JOIN Facilities USING (facid)) AS x
GROUP BY x.name
HAVING ROUND(SUM(CASE
WHEN memid = 0 THEN (x.guestcost * x.slots)
WHEN memid <> 0 THEN (x.membercost * x.slots) END), 2) < 1000
ORDER BY total_revenue; 
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order.

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT (m1.surname || ', ' || m1.firstname) AS member_name, 
        (m2.surname || ', ' || m2.firstname) AS referring_member
FROM Members m1
INNER JOIN Members m2
ON m1.recommendedby = m2.memid
ORDER BY member_name;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q12: Find the facilities with their usage by member, but not guests.

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT 
            facid, 
            COUNT(bookid) AS num_bookings
        FROM Bookings
            WHERE memid <> 0
            GROUP BY facid
            ORDER BY num_bookings DESC;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

Q13: Find the facilities usage by month, but not guests.

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT 
	facid, 
	(CASE
    	WHEN starttime LIKE '2012-01%' THEN '1January' 
     	WHEN starttime LIKE '2012-02%' THEN '2February'
     	WHEN starttime LIKE '2012-03%' THEN '3March'
     	WHEN starttime LIKE '2012-04%' THEN '4April'
     	WHEN starttime LIKE '2012-05%' THEN '5May'
     	WHEN starttime LIKE '2012-06%' THEN '6June'
     	WHEN starttime LIKE '2012-07%' THEN '7July'
    	WHEN starttime LIKE '2012-08%' THEN '8August'
     	WHEN starttime LIKE '2012-09%' THEN '9September'
     	WHEN starttime LIKE '2012-10%' THEN '10October'
     	WHEN starttime LIKE '2012-11%' THEN '11November'
     ELSE '12December' END) AS month, 
     COUNT(bookid) AS num_bookings
FROM Bookings
WHERE memid <> 0
GROUP BY month, facid
ORDER BY facid, month;
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
